```python
import sys
sys.path.append(r'../script/')
from protein_composition import *
import pandas as pd
import numpy as np
import re
import os
import matplotlib.pyplot as plt
import seaborn as sns
# from scipy import stats
```

```python
# zy:test
strain='BW25113'
protein_expression_mass_norm_outfile='../analysis_result/initial_data/BW25113/%s_exp_onecell.json'%strain
protein_expression_mass_norm=json_load(protein_expression_mass_norm_outfile) 
pathway_gene_data = json_load('../analysis_result/pathway_protein/iML1515_pathway_gene.json')  # 存在一个基因对应多个pathway
# pathway_id:pathway_name
pathway_id_name = {}
for value in pathway_gene_data.values():
    pathway_id = value['pathway_id']
    if pathway_id not in pathway_id_name.keys():
        # pathway_id_name["pathway_id"] = pathway_id
        # pathway_id_name["pathway_name"] = value["pathway_name"]
        pathway_id_name[pathway_id] = value["pathway_name"]


print(pathway_id_name)
```

```python
protein_pathway_data = {}
for key_protein, value1 in protein_expression_mass_norm.items():
    protein_pathway_data[key_protein] = {}
    
    for key2_gene, value2 in list(value1.items()):
        # 获取同一基因对应的所有pathway_id
        pid_list = []
        for key1 in list(pathway_gene_data.keys()):
            pid = key1.split("_")[0]
            pathway_gene = key1.split("_")[1]
            if pathway_gene == key2_gene and pid not in pid_list:
                # print(pid,"**")
                pid_list.append(pid)
        # if len(pid_list)>2:     # test
        #     print(key2_gene + "对应pathway_id个数：", len(pid_list))
        for g in pid_list:
            protein_pathway_data[key_protein][g] = value2
```

```python
protein_expression_mass_norm_df=pd.DataFrame(protein_pathway_data) 
protein_expression_mass_norm_df.to_csv('../analysis_result/pathway_protein/pathway_protein.csv')
print(protein_expression_mass_norm_df.shape)
protein_expression_mass_norm_df.head(10)

```

```python
pathway_protein_df = pd.DataFrame()
columns = protein_expression_mass_norm_df.columns.tolist()
for index, col in protein_expression_mass_norm_df.iterrows():
    for i in columns:
        # print(i)
        # print(col[i])
        pathway_protein_df.loc[index+"_"+i, 'value'] = col[i]
        pathway_protein_df.loc[index+"_"+i, 'pathway_id'] = index
        pathway_protein_df.loc[index+"_"+i, 'pathway_name'] = index+":"+ pathway_id_name[index]

print(pathway_protein_df.shape)
pathway_protein_df.to_csv('../analysis_result/pathway_protein/data_for_png.csv')
pathway_protein_df
```

```python
twoamino = pathway_protein_df # .sort_index(axis = 1,ascending = True)
# pathway_protein_df = pd.read_csv('../analysis_result/pathway_protein/pathway_protein.csv', index_col=0)  
pngname = '../analysis_result/pathway_protein/pathway_protein.png'

order_list = []     # x轴顺序
for n in range(1,39):
    order_list.append("g"+ str(n))

legend_labels = []   # 图例注释信息
for key,value in pathway_id_name.items():
    legend_labels.append(key+":"+value)
    
plt.figure(figsize=(25, 10)) 
plt.tick_params(labelsize=18)
sns.boxplot(data=twoamino, x="pathway_id", y="value",order=order_list, hue="pathway_name", hue_order=legend_labels, dodge=False) 
plt.legend(title = "Pathway_Name", title_fontsize = "15", loc = 2, bbox_to_anchor = (1,1))  
plt.ylabel("Value",fontsize=22)
plt.xlabel("Pathway_ID", fontsize=22)
plt.savefig(pngname,dpi =500,bbox_inches='tight')

plt.show()
```

## zy:7种菌株与各自表达值为1，随机值的均值比较

```python
import  matplotlib.pyplot as plt  
strain_list=["W3110", "Yeast_single_cell", "Corynebacterium_RNA_seq", "MG1655", "BW25113", "DH1", "Bacillus"] 
for strain in strain_list:
    amino_composition_norm_onecell_df_outfile='../analysis_result/data_1/%s/amino_composition_proteome_by_condition_%s.csv'%(strain,strain)
    amino_composition_norm_onecell_df=pd.read_csv(amino_composition_norm_onecell_df_outfile,index_col=0)
    twoamino1=amino_composition_norm_onecell_df.sort_index(axis = 1,ascending = True)
    twoamino1_mean=np.mean(twoamino1)

    amino_composition_norm_onecell_df_outfile='../analysis_result/random_data/%s/amino_composition_proteome_by_condition_%s.csv'%(strain,strain)
    amino_composition_norm_onecell_df=pd.read_csv(amino_composition_norm_onecell_df_outfile,index_col=0)
    twoamino3=amino_composition_norm_onecell_df.sort_index(axis = 1,ascending = True)
    twoamino3_mean=np.mean(twoamino3)

    amino_composition_norm_onecell_df_outfile='../analysis_result/initial_data/%s/amino_composition_proteome_by_condition_%s.csv'%(strain,strain)
    amino_composition_norm_onecell_df=pd.read_csv(amino_composition_norm_onecell_df_outfile,index_col=0)
    twoamino2=amino_composition_norm_onecell_df.sort_index(axis = 1,ascending = True)
    twoamino2_mean=np.mean(twoamino2)

    pro_mean=pd.concat([twoamino1_mean,twoamino3_mean, twoamino2_mean],axis=1)
    pro_mean.columns=[strain +'_1',strain +'_random',strain]
    pro_mean.name='Amino'
    pro_mean=pro_mean.sort_index(axis = 0,ascending = True)

    pngname='../analysis_result/article_figures/protein_amino_acid_composition_mean_%s.png'%strain
    plt.figure(figsize=(26, 12)) 
    plt.tick_params(labelsize=20)
    #pro_mean.plot.bar(figsize=(25, 10))
    ind = np.arange(20)                # the x locations for the groups
    plt.xlim(-.2, 20)
    width = 0.15
    #plt.rcParams['font.style'] ='italic' # 'normal'#
    plt.bar(ind,pro_mean[strain +'_1'],width,color = 'blue',label = strain +'_1')  
    plt.bar(ind+2*width,pro_mean[strain +'_random'],width,color = 'red',label = strain +'_random')
    plt.bar(ind+width,pro_mean[strain],width,color = 'green',label = strain) # ind+width adjusts the left start location of the bar.
    
    # plt.bar(ind+3*width,pro_mean['MG1655'],width,color = 'darkorange',label = 'MG1655')
    # plt.bar(ind+4*width,pro_mean['W3110'],width,color = 'violet',label = 'W3110')

    plt.xticks(np.arange(20) + 2*width, list(pro_mean.index),rotation=0)
    #plt.bar(data=pro_mean)
    #sns.barplot(data=pro_mean.T)
    plt.ylabel("Mass ratio (g/g total protein)",fontsize=20)
    plt.xlabel("Amino acid", fontsize=22)  # 我们设置横纵坐标的标题。
    plt.legend(loc="upper right", fontsize=22)
    plt.savefig(pngname,dpi =300,bbox_inches='tight')
    plt.xticks(rotation=0)
    plt.show()
```

## zy:总氨基酸的质量比

```python
strain_list=["W3110", "Yeast_single_cell", "Corynebacterium_RNA_seq", "MG1655", "BW25113", "DH1", "Bacillus"]
for strain in strain_list:
    amino_composition_norm_onecell_df_outfile='../analysis_result/initial_data/%s/amino_composition_proteome_by_condition_%s.csv'%(strain,strain)
    amino_composition_norm_onecell_df=pd.read_csv(amino_composition_norm_onecell_df_outfile,index_col=0)

    print(strain)
    amino_composition_norm_onecell_df=amino_composition_norm_onecell_df.sort_index(axis = 1,ascending = True)
    amino_list=amino_composition_norm_onecell_df.columns.values.tolist()
    amino_mass_norm_stat_std=pd.DataFrame()
                
    for each_i in amino_list:
        data=amino_composition_norm_onecell_df[each_i]
        amino_mass_norm_stat_std.loc[each_i,'mean']=np.mean(data)
        amino_mass_norm_stat_std.loc[each_i,'sd']=np.std(data)
        amino_mass_norm_stat_std.loc[each_i,'sd/mean']=np.std(data)/np.mean(data)
        
    print(amino_mass_norm_stat_std)
    print("**************************")
```
